```python
# Load libraries
import pandas as pd
import matplotlib.pyplot as plt
import squarify
import numpy as np
```

```python
# Load dataset
df = pd.read_csv('data/expenditures.txt', sep='\t')

# View the dataframe
df
```

```python
# Build a tree map

#Trim dataframe to only contain data from 2008
df_2008 = df.loc[df['year']==2008]

# Build tree map
squarify.plot(sizes=df_2008['expenditure'], label=df_2008['category'], alpha=.8)
plt.axis('off')
plt.title('2008 Expenditures by Category')
plt.show()
```

```python
# Group expenditure by year
df_grouped = df.groupby('year').sum()

# Reset the index
df_grouped = df_grouped.reset_index()
```

```python
# Create an area chart
plt.fill_between(df_grouped['year'], df_grouped['expenditure'], color="skyblue", alpha=0.4)
plt.plot(df_grouped['year'], df_grouped['expenditure'], color="Slateblue", alpha=0.6, linewidth=2)
plt.box(False)
plt.title('Expenditures by Year', loc='left', fontsize=15, color='grey')
plt.xlabel('Year', fontsize=12, color='grey')
plt.ylabel('Expenditure', fontsize=12, color='grey')
plt.tick_params(axis='x', colors='grey')
plt.tick_params(axis='y', colors='grey')
plt.show()
```

```python
# Obtain list of categories
for i in df['category'].unique():
    print(i)
```

```python
# Create dataframes for each category
df_food = df.loc[df['category']=='Food']
df_alcohol = df.loc[df['category']=='Alcoholic Beverages']
df_housing = df.loc[df['category']=='Housing']
df_apparel = df.loc[df['category']=='Apparel']
df_transportation = df.loc[df['category']=='Transportation']
df_healthcare = df.loc[df['category']=='Healthcare']
df_entertainment = df.loc[df['category']=='Entertainment']
df_personalcare = df.loc[df['category']=='Personal Care']
df_reading = df.loc[df['category']=='Reading']
df_education = df.loc[df['category']=='Education']
df_tobacco = df.loc[df['category']=='Tobacco Products']
df_miscellaneous = df.loc[df['category']=='Miscellaneous']
df_cash = df.loc[df['category']=='Cash Contributions']
df_personalinsurance = df.loc[df['category']=='Personal Insurance']
```

```python
plt.stackplot(df_food['year'], df_food['expenditure'], df_alcohol['expenditure'], df_housing['expenditure'], 
              df_apparel['expenditure'], df_transportation['expenditure'], df_healthcare['expenditure'], 
              df_entertainment['expenditure'], df_personalcare['expenditure'], df_education['expenditure'], 
              df_tobacco['expenditure'], df_miscellaneous['expenditure'], df_cash['expenditure'], 
              df_personalinsurance['expenditure'], labels=['Food', 'Alcoholic Beverages', 'Housing', 'Apparel', 
                                                           'Transporation', 'Healthcare', 'Entertainment', 'Personal Care',
                                                          'Reading', 'Education', 'Tobacco Products', 'Miscellaneous', 
                                                          'Cash Contributions', 'Personal Insurance'])
plt.legend(loc='upper left', fontsize=12)
plt.title('Expenditures by Year and Category', loc='left', fontsize=15, color='grey')
plt.xlabel('Year', fontsize=12, color='grey')
plt.ylabel('Expenditure', fontsize=12, color='grey')
plt.tick_params(axis='x', colors='grey')
plt.tick_params(axis='y', colors='grey')
plt.rcParams["figure.figsize"] = (18,10)
```
