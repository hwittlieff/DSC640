```python
# Load libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_excel('data/world-population.xlsm', index_col=0)

# Reset index
df = df.reset_index()

# View the dataframe
df
```

```python
# Build line chart
plt.plot(df['Year'], df['Population'])
plt.title('World Population by Year')
plt.xlabel('Year')
plt.ylabel('Population (in Billions)')
plt.ylim([0, 7000000000])
plt.show()
```

```python
# Build step chart
plt.step(df['Year'], df['Population'])
plt.title('World Population by Year')
plt.xlabel('Year')
plt.ylabel('Population (in Billions)')
plt.ylim([0, 7000000000])
plt.show()
```
