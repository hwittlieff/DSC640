```python
# Load libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Load the costco dataset
df_costco = pd.read_csv('Data/costcos-geocoded.csv')

# View the dataframe
df_costco
```

```python
# Load the ppg2008 dataset
df_bb = pd.read_csv('Data/ppg2008.csv')
df_bb = df_bb.set_index('Name  ')


# view the dataframe
df_bb
```

# Heatmap

```python
# Normalize data
df_bb_norm = (df_bb-df_bb.mean())/df_bb.std()
```

```python
# Build a heatmap
sns.set(rc = {'figure.figsize':(30,30)})
sns.set(font_scale = 2)
sns.heatmap(df_bb_norm, cmap="mako", annot=df_bb, cbar=False).set(title='2008 NBA Performance')
plt.show()
```

# Spatial Chart

```python
# Load libraries
from mpl_toolkits.basemap import Basemap
```

```python
df_costco
```

```python
fig = plt.figure(figsize=(8, 8))
m = Basemap(projection='lcc', resolution=None,
            width=8E6, height=8E6, 
            lat_0=45, lon_0=-100,)
m.etopo(scale=0.5, alpha=0.5)

for index, row in df_costco.iterrows():
    City = row['City']
    Latitude = row['Latitude']
    Longitude = row['Longitude']
    x, y = m(Longitude, Latitude)
    plt.plot(x, y, 'ok', markersize=5)

plt.title('USA Costco Locations')
```
