```python
# Load libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Bullet Chart

```python
# Create staff per student column
df['Staff Per Student'] = 1/df['pupil_staff_ratio']
```

```python
def bullet_chart(df,color_code=False):
  y_pos = np.arange(len(df.index))

  #assign coloring
  df["col"]="indigo"
  if (color_code==True):
    for i in y_pos:
      if(df["current value"][i]>=df["stretch goal"][i]):
        df["col"][i]="green"
      elif(df["current value"][i]>=df["base goal"][i]):
        df["col"][i]="gold"
      else:
        df["col"][i]="lightcoral"

  #Initialize plot
  fig, ax = plt.subplots()
  ax.barh(y_pos, df["stretch goal"], height=0.5, align='center', color='thistle', label="stretch goal")
  ax.barh(y_pos, df["base goal"], height=0.5, align='center', color='mediumorchid', label = "base goal")
  ax.barh(y_pos, df["current value"], height=0.2, align='center',color=df["col"])
  ax.set_yticklabels(df["labels"])
  ax.set_yticks(y_pos)
  ax.invert_yaxis()
  ax.set_title('Staff Per Student by State')
    
  #add data labels
  for i in y_pos:
    ax.text(df["current value"][i], i+0.05, df["current value"][i])

  #add legend and format borders
  plt.legend(loc=(0.35,1.0))
  ax.spines['left'].set_linewidth(0.2)
  ax.spines['bottom'].set_linewidth(0)
  ax.spines['right'].set_linewidth(0)
  ax.spines['top'].set_linewidth(0)
  fig.subplots_adjust(left=0.2, top=0.8)

  return fig
```

```python
# Create df
df_bullet=df
df_bullet['labels'] = df_bullet['state']
df_bullet['base goal'] = .05
df_bullet['stretch goal'] = .1
df_bullet['current value'] = df_bullet['Staff Per Student']
```

```python
bullet_chart(df_bullet.loc[:5, :], 'Test')
```
