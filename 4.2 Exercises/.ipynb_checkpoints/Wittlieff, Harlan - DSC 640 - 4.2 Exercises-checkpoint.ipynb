```python
# Load libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv('Data/crimerates-by-state-2005.csv')

# View the dataframe
df
```

```python
# Drop the USA & District of Columbia Rows
df = df.drop([0, 9])

# Reset axis
df = df.reset_index()
```

```python
# Create a scatterplot
plt.rcParams["figure.figsize"] = (15,10)
plt.scatter(df['robbery'], df['murder'], color="skyblue", alpha=0.4)
plt.box(False)
plt.title('Robberies vs. Murders in the U.S.', loc='left', fontsize=20, color='grey')
plt.tick_params(axis='x', colors='grey')
plt.tick_params(axis='y', colors='grey')
plt.xlabel('Robberies per 100,000 Population', fontsize=15, color='grey')
plt.ylabel('Murders per 100,000 Population', fontsize=15, color='grey')
for i, txt in enumerate(df['state']):
    plt.annotate(txt, (df['robbery'][i], df['murder'][i]), color='grey', fontsize=15)
```

```python
# Create normalized population
p_normal = df['population']/df['population'].max()
```

```python
# Create a bubble plot
plt.rcParams["figure.figsize"] = (15,10)
plt.scatter(df['robbery'], df['murder'], s=p_normal*2000, color="skyblue", alpha=0.4)
plt.box(False)
plt.title('Robberies vs. Murders in the U.S.', loc='left', fontsize=20, color='grey')
plt.tick_params(axis='x', colors='grey')
plt.tick_params(axis='y', colors='grey')
plt.xlabel('Robberies per 100,000 Population', fontsize=15, color='grey')
plt.ylabel('Murders per 100,000 Population', fontsize=15, color='grey')
for i, txt in enumerate(df['state']):
    plt.annotate(txt, (df['robbery'][i], df['murder'][i]), color='grey', fontsize=15)
```

```python
ax = sns.kdeplot(np.array(df['murder']))
ax.set(xlabel='Murders per 100,000 Population')
sns.despine()
ax.set(title='U.S. Murder Rate by State')
```
